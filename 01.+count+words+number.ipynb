* 1.把所有列名都知道，放入一个列表中
* 2.将所有相同的名称拿出来进行统计个数
    (1)遍历此行中
* 3.创建相应数量的字典把他们进行存储

```python
import pandas as pd
name_list = []
data = pd.read_csv("final_test.csv")
data.columns
```

```python
for i in data["Name"]:
    if i not in name_list:
        name_list.append(i)
        
print(name_list)
```

```python
#将每个不同的项目数据储存到相应的文档中
#只要执行一遍
files_name = []
def file_collect(fold_name, item):
    num = 0
    for j in name_list:
        file_name = str(j)+"("+item +")"+".txt"
        files_name.append(file_name)
        for k in data["Name"]:
            if j == k:
                with open(fold_name+"/"+file_name,"a") as f:
                    content = str(data[item][num])
                    num += 1
                    f.write(content+"\n")
            else:
                f.close()
                continue
```

```python
#进行数据的操作计算
import sys,re
import codecs

tenNum=[]
result = []
lineCounts = 0
easyWords = []
distone={}
def openFile(oldFolderName, name):
        global lineCounts
        
        #打开一个文档
        fr = codecs.open(oldFolderName+"/"+name,"r",encoding= u'utf-8',errors='ignore')
        for line in fr.readlines():
                lineCounts += 1
                line = re.sub("\W"," ", line)
  
                lineone = line.lower().split() 
                #print(line)
                #print(lineone)
                for keyone in lineone:
                    
                    if not distone.get(keyone):
                        distone[keyone]=1
                    else:
                        distone[keyone]+=1
                                  #print(distone) 用字典的方式将出现的次数进行存储
        fr.close()
        print(type(distone))

def orderNum():
        #根据每个单词出现的次数来选出前100
        global tenNum
        for v in distone.values():
                 if v not in tenNum:
                         tenNum.append(v)
 
        tenNum.sort()
        tenNum=tenNum[-100:]
 
 
def orderData(newFolderName, name):
        global distone
        distone_item = sorted(distone.items(),key=lambda d:d[1],reverse = True)
 
        fw = codecs.open(newFolderName+"/"+name,"w",encoding= u'utf-8',errors='ignore')
        for i in distone_item:
             if i[1] in tenNum:
                    print(i)
                    fw.write(str(i)+"\n")
        fw.write(str(lineCounts))
        fw.close()    
```

```python
items_name = ["IssueDetails", "ResolutionSteps", "History"]
file_collect("IssueDetails", "IssueDetails")
file_collect("ResolutionSteps","ResolutionSteps")
file_collect("History","History")
#print(files_name)

def main():
    for x in files_name:
        if x != "nan(IssueDetails).txt":
            openFile("IssueDetails", x)
            orderNum()
            orderData("newIssueDetails", x)
    
```

```python
if __name__ =="__main__":  
    main()
    
```
